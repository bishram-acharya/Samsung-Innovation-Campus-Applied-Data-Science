```python
import numpy as np


x1 = np.random.rand(124, 1)
x2 = np.random.rand(124, 1)
x3 = np.random.rand(124, 1)
x4 = np.random.rand(124, 1)
x5 = np.random.rand(124, 1)


vectors = [x1, x2, x3, x4, x5]


closest_pair = (x1, x2)
smallest_value = np.dot(x1.T, x2) / (np.linalg.norm(x1) * np.linalg.norm(x2))


for i in range(len(vectors)):
    for j in range(i + 1, len(vectors)):
        vec1 = vectors[i]
        vec2 = vectors[j]
        cos_sim = np.dot(vec1.T, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2))
        if cos_sim < smallest_value:
            smallest_value = cos_sim
            closest_pair = (vec1, vec2)


smallest_value
closest_pair
```
