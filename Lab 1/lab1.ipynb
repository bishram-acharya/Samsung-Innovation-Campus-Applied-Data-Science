#### Who was the oldest/youngest passenger 

```python

low=22
high=38
low_index=0
high_index=1
for i, value in enumerate(df["Age"]):
    if value<low:
        low_index=i
Passenger_id_low, age_low= low_index+1, df["Age"][low_index]
for i, value in enumerate(df["Age"]):
    if value>high:
        high_index=i
Passenger_id_high, age_high = high_index+1,df["Age"][high_index]
print(f'The youngest passenger was ID {Passenger_id_low} and age was {age_low} \n The oldest passenger was ID {Passenger_id_high} and age was {age_high}')
```

```python
survived_youngest=df.iloc[low_index][1]
survived_oldest=df.iloc[high_index][1]
survived_youngest,survived_oldest
```

```python
df_surv=df[df["Survived"]==1].reset_index(drop=True)
df_dead=df[df["Survived"]==0].reset_index(drop=True)
```

#### Who was the oldest/youngest passenger  to die

```python

low=22
high=35
low_index=0
high_index=4
for i, value in enumerate(df_dead["Age"]):
    if value<low:
        low_index=i
Passenger_id_low, age_low= low_index+1, df_dead["Age"][low_index]
for i, value in enumerate(df_dead["Age"]):
    if value>high:
        high_index=i
Passenger_id_high, age_high = high_index+1,df_dead["Age"][high_index]
print(f'The youngest passenger to die was ID {Passenger_id_low} and age was {age_low} \n The oldest passenger to die was ID {Passenger_id_high} and age was {age_high}')
```

#### Who was the oldest/youngest passenger  to survive


```python
low=26
high=38
low_index=1
high_index=0
for i, value in enumerate(df_surv["Age"]):
    if value<low:
        low_index=i
Passenger_id_low, age_low= low_index+1, df_surv["Age"][low_index]
for i, value in enumerate(df_surv["Age"]):
    if value>high:
        high_index=i
Passenger_id_high, age_high = high_index+1,df_surv["Age"][high_index]
print(f'The youngest passenger to survive was ID {Passenger_id_low} and age was {age_low} \n The oldest passenger to survive was ID {Passenger_id_high} and age was {age_high}')
```
